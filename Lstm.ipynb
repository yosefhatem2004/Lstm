```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

weights = {
    'f': {'w': 0.5, 'wh': 0.1, 'b': 0},
    'i': {'w': 0.6, 'wh': 0.2, 'b': 0},
    'c': {'w': 0.7, 'wh': 0.3, 'b': 0},
    'o': {'w': 0.8, 'wh': 0.4, 'b': 0}
}

h_t = 0
c_t = 0
X = [1, 2, 3]

print(f"{'Step':<10} | {'f_t':<8} | {'i_t':<8} | {'C_tilde':<8} | {'C_t':<8} | {'h_t':<8}")
print("-" * 65)

for t, x_t in enumerate(X, 1):
    f_t = sigmoid(weights['f']['w'] * x_t + weights['f']['wh'] * h_t + weights['f']['b'])
    i_t = sigmoid(weights['i']['w'] * x_t + weights['i']['wh'] * h_t + weights['i']['b'])
    c_tilde = np.tanh(weights['c']['w'] * x_t + weights['c']['wh'] * h_t + weights['c']['b'])
    c_t = (f_t * c_t) + (i_t * c_tilde)
    o_t = sigmoid(weights['o']['w'] * x_t + weights['o']['wh'] * h_t + weights['o']['b'])
    h_t = o_t * np.tanh(c_t)

    print(f"t={t} (x={x_t}) | {f_t:.3f}    | {i_t:.3f}    | {c_tilde:.3f}    | {c_t:.3f}    | {h_t:.3f}")

x_4 = 4
f_4 = sigmoid(weights['f']['w'] * x_4 + weights['f']['wh'] * h_t + weights['f']['b'])
i_4 = sigmoid(weights['i']['w'] * x_4 + weights['i']['wh'] * h_t + weights['i']['b'])
c_tilde_4 = np.tanh(weights['c']['w'] * x_4 + weights['c']['wh'] * h_t + weights['c']['b'])
c_4 = (f_4 * c_t) + (i_4 * c_tilde_4)
o_4 = sigmoid(weights['o']['w'] * x_4 + weights['o']['wh'] * h_t + weights['o']['b'])
h_4 = o_4 * np.tanh(c_4)

w_y, b_y = 4, 0
y_hat = w_y * h_4 + b_y

print("-" * 65)
print(f"Final Hidden State (h4): {h_4:.3f}")
print(f"Final Prediction (y_hat): {y_hat:.3f}")
```
